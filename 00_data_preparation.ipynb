```python
import pandas as pd
from IPython.display import display

# DATA LOAD + QUICK EDA
TRAIN_PATH = "train.parquet"

train = pd.read_parquet(TRAIN_PATH)

print(f"Train shape: {train.shape}")
display(train.head(5))

# ---- Schema (colonnes, types, missing) ----
schema = pd.DataFrame({
    "column": train.columns,
    "dtype": train.dtypes.astype(str).values,
    "missing_pct": (train.isna().mean() * 100).round(2).values,
    "n_unique": train.nunique(dropna=True).values
}).sort_values("missing_pct", ascending=False)

display(schema)

# ---- Top pages ----
if "page" in train.columns:
    display(train["page"].value_counts(dropna=False).head(20).rename_axis("page").reset_index(name="count"))
```
